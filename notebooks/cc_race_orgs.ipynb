```python
def get_closest_match(x, choices):
    name = process.extractOne(x, choices)[0] 
    score = process.extractOne(x, choices)[1]

    if score > 90:
        return name
    else:
        return x
```

```python
# apply function to the 'NAME' column, set the result to two separate columns
all_cc_org['closest_match'] = all_cc_org['NAME'].apply(lambda x: get_closest_match(x, org_key))
```

```python
def normalize_name(name):
    if ',' in name:
        last_name, first_name = name.split(', ')
        return f"{first_name} {last_name}"
    else:
        return name
```

```python
all_cc_org['normalized_mem'] = all_cc_org['RECIPNAME'].apply(lambda x: normalize_name(x))
```

```python
def fuzzy_merge(df1, df2, key1, key2, threshold=90):
    matched_names = []
    
    # Iterate over the names in df1
    for name in df1[key1]:
        # ExtractOne returns (best_match, score, index)
        match, score, _ = process.extractOne(name, df2[key2], scorer=fuzz.token_sort_ratio)
        if score >= threshold:
            matched_names.append((name, match, score))
        else:
            matched_names.append((name, None, None))
    
    # Create a dataframe of matches
    return pd.DataFrame(matched_names, columns=[key1, key2, 'score'])
```

```python
fuzzy_matched_df = fuzzy_merge(faction_compare, all_cc_org, 'VotePersonName', 'normalized_mem', threshold=70)
```
